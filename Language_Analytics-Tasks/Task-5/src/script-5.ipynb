### 1. Import the packages.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

### 2.Load the data.
It loads in only the columns "task_name" and "emissions", which are the only ones needed to complete the analysis.

```python
a1 = pd.read_csv("../in/emissions-A-1.csv", usecols=['task_name', 'emissions'])
a2_log = pd.read_csv("../in/emissions-A-2-log.csv", usecols=['task_name', 'emissions'])
a2_neural = pd.read_csv("../in/emissions-A-2-neural.csv", usecols=['task_name', 'emissions'])
a3 = pd.read_csv("../in/emissions-A-3.csv", usecols=['task_name', 'emissions'])
a4 = pd.read_csv("../in/emissions-A-4.csv", usecols=['task_name', 'emissions'])

# Creates a dataframe with emission values for assignment 2:
a2 = pd.concat([a2_log, a2_neural], ignore_index=True)
```

### 3. Plot the total amount of emissions across all assignments.

3.1. Calculate the total amount of emissions produced by running each Python script:

```python
total_a1 = sum(a1["emissions"])
total_a2_log = sum(a2_log["emissions"])
total_a2_neural = sum(a2_neural["emissions"])
total_a2 = total_a2_log + total_a2_neural
total_a3 = sum(a3["emissions"])
total_a4 = sum(a4["emissions"])
```

3.2. Create a dataframe with all these values:

```python
total_values = [total_a1, total_a2, total_a3, total_a4]
assignment_labels = ["Assignment 1", "Assignment 2", "Assignment 3", "Assignment 4"]

df_1 = pd.DataFrame({
    'Assignment': assignment_labels,
    'Total Values': total_values
})
```

3.3. Plot the data and save the figure:

```python
plt.figure(figsize=(8, 4))
plt.bar(df_1['Assignment'], df_1['Total Values'])
plt.title('Comparison of Total CO₂eq across Assignments')
plt.xlabel('Assignments')
plt.ylabel('Total CO₂eq')
plt.grid(True, which='both', axis='y', linewidth=0.5)

# Save the plot:
plt.savefig("../out/total_emissions.png")

# Show the plot:
plt.show()
```

### 4. Plot the emission data per subtask for each assignment.

```python
# Creates subplots and size of the figure:
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# Plotting Assignment 1
axs[0, 0].bar(a1['task_name'], a1['emissions'], color='#1f77b4')
axs[0, 0].set_title('Assignment 1')
axs[0, 0].set_xlabel('Task')
axs[0, 0].set_ylabel('CO₂eq')
axs[0, 0].grid(True, which='both', axis='y', linewidth=0.5)
axs[0, 0].tick_params(axis='x', rotation=45)

# Plotting Assignment 2
axs[0, 1].bar(a2['task_name'], a2['emissions'], color='#a9cae2')
axs[0, 1].set_title('Assignment 2')
axs[0, 1].set_xlabel('Task')
axs[0, 1].set_ylabel('CO₂eq')
axs[0, 1].grid(True, which='both', axis='y', linewidth=0.5)
axs[0, 1].tick_params(axis='x', rotation=45)

# Plotting Assignment 3
axs[1, 0].bar(a3['task_name'], a3['emissions'], color='#7aa9d3')
axs[1, 0].set_title('Assignment 3')
axs[1, 0].set_xlabel('Task')
axs[1, 0].set_ylabel('CO₂eq')
axs[1, 0].grid(True, which='both', axis='y', linewidth=0.5)
axs[1, 0].tick_params(axis='x', rotation=45)

# Plotting Assignment 4
axs[1, 1].bar(a4['task_name'], a4['emissions'], color='#4b88c3')
axs[1, 1].set_title('Assignment 4')
axs[1, 1].set_xlabel('Task')
axs[1, 1].set_ylabel('CO₂eq')
axs[1, 1].grid(True, which='both', axis='y', linewidth=0.5)
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()

plt.savefig("../out/emissions_subtasks.png")

plt.show()
```
